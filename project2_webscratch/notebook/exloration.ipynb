```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Start browser
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Open your page
url = "https://www.eu.lululemon.com/en-lu/c/mens/collections/we-made-too-much?pmid=markdownPromotion&prefn1=activity&prefv1=Workout%7CRunning&prefn2=size&prefv2=XS%7CS&srule=Price%20Ascending"
driver.get(url)

for _ in range(5):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)


print(driver.title)
```

```python
from selenium.webdriver.common.by import By

products = driver.find_elements(By.CSS_SELECTOR, "a.search-results-product-name")

print(len(products))  # how many found
```

```python
data = []

for p in products:
    name = p.text.strip()
    link = p.get_attribute("href")
    price = p.get_attribute("data-product-price")
    
    data.append({
        "name": name,
        "price": price,
        "link": link
    })

keyword = ["short", "tank"]
for item in data[:]:
    if any(kw in item['name'].lower() for kw in keyword):
        print(f"{item['name']},=======EUR=== {item['price']}, \n{item['link']}")
```
